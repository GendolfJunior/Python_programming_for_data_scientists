```python
def find_min_max(row):
    data = row[['POPESTIMATE2010',
                  'POPESTIMATE2011',
                  'POPESTIMATE2012',
                  'POPESTIMATE2013',
                  'POPESTIMATE2014',
                  'POPESTIMATE2015']]
    row['min']=np.min(data)
    row['max']=np.max(data)
    return row
```

### Pandas GroupBy

```python
df = pd.read_csv('data/census.csv')
```

```python
df = df[df['SUMLEV']==50]
```

```python
df.head()
```

```python
df['STNAME'].unique()
```

```python
%%timeit -n 10
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Average population in ' + state + ' is ' + str(avg))
```

```python
df.groupby('STNAME')
```

```python
%%timeit -n 10
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Average population in ' + group + ' is ' + str(avg))
```

```python
def func(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2
```
