### Unsupervised task

```python
# one of dimensionality reduction models is Principal Component Analysis (PCA)
from sklearn.decomposition import PCA
```

```python
# we start always with the same steps.
# 1. you import the model
# 2. you initialize the model
# n_components - is about how many spaces, below is 2 dimensional space
model = PCA(n_components=2)
```

```python
model.fit(X_iris)
```

```python
X_2d = model.transform(X_iris)
```

```python
model.transform?
```

```python
X_2d
```

```python
iris['PCA1'] = X_2d[:, 0]
iris['PCA2'] = X_2d[:, 1]
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False)
```

```python
sns.set()
sns.pairplot(iris, hue='species')
```

```python
from sklearn.mixture import GMM
```

```python
model = GMM(n_components = 3)
```

```python
model.fit(X_iris)
```

```python
y_gmm = model.predict(X_iris)
```

```python
iris['cluster'] = y_gmm
```

```python
sns.lmplot('PCA1', 'PCA2', data=iris, hue='species', col='cluster', fit_reg=False)
```
